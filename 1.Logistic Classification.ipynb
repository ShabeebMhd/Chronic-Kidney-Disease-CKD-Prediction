```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('CKD.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b','sg_c', 'sg_d', 'sg_e', 'rbc_normal',
                     'pc_normal','pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent
```

```python
dependent=dataset[['classification_yes']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.2, random_state=0)
```

```python

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train_scaled=sc.fit_transform(x_train)
x_test_scaled=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
classifier=LogisticRegression()

param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}

grid=GridSearchCV(classifier, param_grid, cv=5, scoring='f1_weighted',refit=True, n_jobs=-1)
grid.fit(x_train_scaled,y_train)
```

```python
re=grid.cv_results_
print("The cofution matrics for the best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_prediction=grid.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,grid_prediction)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
clf_report=(classification_report(y_test, grid_prediction))
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(x_test_scaled)[:,1])
```
