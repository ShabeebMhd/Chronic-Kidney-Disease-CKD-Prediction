```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv','wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal',
                    'pc_normal','pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes','appet_yes', 'pe_yes', 'ane_yes']]
```

```python
dependent=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)   
x_test_scaled = scaler.transform(x_test)         
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB, MultinomialNB,BernoulliNB,ComplementNB, CategoricalNB

pipeline = Pipeline([
('classifier', GaussianNB()) # placeholder, will be replaced by param_grid
])

param_grid = [
{
'classifier': [GaussianNB()],
'classifier__var_smoothing': [1e-12, 1e-11, 1e-10, 1e-9, 1e-8, 1e-7]
},
{
'classifier': [MultinomialNB()],
'classifier__alpha': [0.01, 0.1, 0.5, 1.0, 2.0, 5.0],
'classifier__fit_prior': [True, False]
},
{
'classifier': [BernoulliNB()],
'classifier__alpha': [0.01, 0.1, 0.5, 1.0, 2.0],
'classifier__binarize': [0.0, 0.5, 1.0, None],
'classifier__fit_prior': [True, False]
},
{
'classifier': [ComplementNB()],
'classifier__alpha': [0.01, 0.1, 0.5, 1.0, 2.0],
'classifier__fit_prior': [True, False],
'classifier__norm': [True, False]
},
{
'classifier': [CategoricalNB()],
'classifier__alpha': [0.01, 0.1, 0.5, 1.0, 2.0],
'classifier__fit_prior': [True, False]
}
]

grid = GridSearchCV(pipeline, param_grid, cv=5,refit=True, verbose=1, n_jobs=-1)
grid.fit(x_train_scaled,y_train)
```

```python
re=grid.cv_results_
print("The cofution matrics for the best algorithm and parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_pred=grid.predict(x_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_pred)
print("Confusion Matrix:\n", cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, grid_pred))
```

```python
from sklearn.metrics import roc_auc_score
y_proba = grid.predict_proba(x_test_scaled)[:, 1]

roc_auc_score(y_test,grid.predict_proba(x_test_scaled)[:,1])
```
